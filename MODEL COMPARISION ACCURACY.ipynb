```python
import matplotlib.pyplot as plt

# Data from your table
models = [
    "Simple CNN", "Simple VGG16", "CNN + SVM (w/o PCA)", "CNN + SVM (w/ PCA)", 
    "CNN + RF (w/o PCA)", "CNN + RF (w/ PCA)", 
    "VGG16 + SVM (w/o PCA)", "VGG16 + SVM (w/ PCA)", 
    "VGG16 + RF (w/o PCA)", "VGG16 + RF (w/ PCA)" 
]
accuracies = [84.51, 78.24, 75.64, 86.72, 86.89, 85.9, 84.8, 82.66, 79.12, 68.62]

# Plotting the bar graph
plt.figure(figsize=(12, 6))
bars = plt.bar(models, accuracies, color='teal', alpha=0.85)

# Adding data labels on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.3, f"{yval:.2f}", 
             ha='center', va='bottom', fontsize=10)

# Customizing labels and title
plt.xlabel("Models", fontsize=13)
plt.ylabel("Accuracy (%)", fontsize=13)
plt.title("Model Accuracy Comparison", fontsize=15, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.ylim(65, 90)  # Adjust y-axis limits for better focus

# Minimal gridlines for reference
plt.grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.7)

plt.tight_layout()
plt.show()
```
